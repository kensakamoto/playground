```python
import numpy as np
n = 5
a = np.array([2, 3, 3, 5, 6, 7, 8])
k = 7
```

```python
def solve():
    lower_bound = 0
    upper_bound = n 
    while upper_bound - lower_bound > 1:
        mid = int((lower_bound + upper_bound) / 2)
        print(mid)
        if a[mid] >= k:
            upper_bound = mid
        else:
            lower_bound =mid
    return upper_bound
```

```python
solve()
```
